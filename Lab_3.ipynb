MULTILAYER PERCEPTRON FOR  DECISION EXCLUSIVE OR task (XOR  task) 

```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
from tensorflow.keras.layers import Dense
adam = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
X = np.array([[2,4],[-2,4],[2,-4], [-2,-4]])
X = X/X.max(axis=0)
y = np.array([[0],[1],[1],[0]])
```

```python
def create_model():
  model = tf.keras.models.Sequential([Dense(2,input_shape=(X.shape[1],), activation='sigmoid'), Dense(1, activation='sigmoid')])
  model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), loss='mse')
  return model
```

```python
model = create_model()
history = model.fit(X, y, epochs=1000)
```

```python
pd.DataFrame(history.history).plot()
plt.title("The error  change depending on the epoch")
```

```python
y_pred = model.predict(X)
y_pred
```
